# Comparing Headlines from CBC and FOX

A simple experiment to show the headlines from CBC and FOX.

Let's start with FOX.

```python
import urllib.request
from bs4 import BeautifulSoup

foxSource = urllib.request.urlopen('https://www.foxnews.com').read()
```

We have the source code, now we're going to look at getting the headlines using Beautiful Soup.

```python
foxSoup = BeautifulSoup(foxSource, 'lxml')
foxArticles = foxSoup.find_all("article")
for foxArticle in foxArticles:
    foxHeader = foxArticle.find("header")
    if foxHeader:
        foxAnchor = foxHeader.find("a")
        print(foxAnchor.string)
```
